## Outliers

We noticed that some dataset values have Salary with 0 USD. 

```python
mask_ss = jobsala.salary_in_usd > 0
jobsala = jobsala[mask_ss]
data.describe()
```

Exploring the data outliers with boxplot.

```python
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)

x = jobsala['salary_in_usd']

plt.figure(figsize = (30,10))

plt.subplot(2,1,1)
sns.boxplot(x)
plt.xlabel("")
plt.title('Salary Distribution', weight=50, size = 30, pad = 20)

plt.subplot(2,1,2)
sns.distplot(x, kde = False)
plt.xlabel('Salary Distribution (in USD)]', size = 25)

plt.show()
```

We have some outliers in salary distribution.

Used IQR; we drop that outliers in the dataset.

```python
sP_Supp = jobsala['salary_in_usd']
IQR = 1.5*(sP_Supp.quantile(.75)-sP_Supp.quantile(.25))
lim_sup = sP_Supp.quantile(.75)+IQR
lim_inf = sP_Supp.quantile(.25)-IQR
```

```python
jobsala.shape
```

```python
sP_clean_Supp = sP_Supp[(sP_Supp >= lim_sup) | (sP_Supp <= lim_inf)]
```

```python
data_job=jobsala.drop(sP_clean_Supp.index)
```

```python
data_job.describe()
```

```python
import warnings
warnings.simplefilter(action="ignore", category=FutureWarning)

x = data_job['salary_in_usd']

plt.figure(figsize = (30,10))

plt.subplot(2,1,1)
sns.boxplot(x)
plt.xlabel("")
plt.title('Salary Distribution', weight=50, size = 30, pad = 20)

plt.subplot(2,1,2)
sns.distplot(x, kde = False)
plt.xlabel('Salary Distribution (in USD)]', size = 25)

plt.show()
```
