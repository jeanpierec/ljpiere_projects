```python
df = pd.read_csv('data/heart/heart.csv')
df.head()
```

```python
import matplotlib.pyplot  as plt
%matplotlib inline
df.boxplot('chol')
plt.title("Boxplot with Chol")
plt.show()
```

```python
import matplotlib.pyplot  as plt
df.boxplot(by = 'sex',column = 'chol')
```

```python
import seaborn as sns
sns.boxplot(x = 'sex', y = 'chol', data = df) 
```

```python
sns.set()
sns.boxplot(x = 'sex', y = 'chol', data = df, width=0.2) 
```

```python
sns.swarmplot(x = 'sex', y = 'chol', data = df, alpha=0.5)
sns.boxplot(x = 'sex', y = 'chol', data = df, width=0.2)
```

```python
sns.boxplot(x = 'sex', y = 'chol', data = df, width=0.2)
sns.swarmplot(x = 'sex', y = 'chol', data = df, alpha=0.5)
plt.title("Colesterol por sexo")
plt.savefig('boxplot.png',dpi=100)
```
