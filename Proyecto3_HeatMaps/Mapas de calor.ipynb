```python
df = pd.read_csv('data/heart/heart.csv')
df.head(12)
```

```python
df.corr()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.imshow(df.corr())
```

```python
corr=df.corr()

fig, ax = plt.subplots()
plt.imshow(corr)
plt.xticks(range(len(corr)), corr.index, rotation=75) #rotacion, inclinacion para palabras en eje x
plt.yticks(range(len(corr)), corr.index)
plt.show()
```

```python
corr=df.corr()

fig, ax = plt.subplots()
plt.imshow(corr,cmap='PuOr') #<- https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
plt.xticks(range(len(corr)), corr.index, rotation=75)
plt.yticks(range(len(corr)), corr.index)
plt.colorbar()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pylab as plt
plt.figure(figsize=(12,10))
sns.heatmap(corr)
sns.heatmap(corr, linewidth=0.2)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pylab as plt
plt.figure(figsize=(12,10))
sns.heatmap(corr, linewidth=0.5,annot=True,cmap="YlGnBu") #<- https://matplotlib.org/3.1.0/tutorials/colors/colormaps.html
plt.show()
```
