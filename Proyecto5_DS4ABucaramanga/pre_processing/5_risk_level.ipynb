```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Determine the risk determine the level risk for each cluster for each cluster in model for 2021 and 2022

## Clustergin model 2021

```python
"""
Load dataframe
"""
GMM_2021 = pd.read_csv("df_clustered_GMM.csv",delimiter="~", encoding = 'latin1' )
GMM_2021.head()
```

```python
"""
Groupby Cluter and delete columns that aren't risk factors
"""
Risk_score_2021 = GMM_2021.groupby("Cluster").mean().reset_index()
Risk_score_2021 = Risk_score_2021.drop(["cod_barrio","cod_mpio","cod_dpto","sexo_persona","edad_c","cod_vereda"],axis=1)
```

```python
"""
calculate percentage for each risk factor
"""

df_pct = Risk_score_2021/Risk_score_2021[Risk_score_2021.columns].sum()
df_pct["Cluster"] = round(df_pct["Cluster"]*10)
```

```python
"""
calculate weighted for each risk factor
"""
df_pct["risk_weight"] = round(df_pct.mean(axis=1),2)
```

```python
"""
Stablish conditions foe each risk level
"""
conditionlist = [
    (df_pct['risk_weight'] == 0.46 ) ,
    (df_pct['risk_weight'] == 0.35 ),
    (df_pct['risk_weight'] == 0.24 ),
    (df_pct['risk_weight'] == 0.20 ),
    (df_pct['risk_weight'] == 0.12) ]
choicelist = ['High', 'High-Mid', 'Mid', 'Mid-Low', 'Low']
```

```python
"""
Create risk level in dataframe
"""
df_pct["risk_level"] = np.select(conditionlist, choicelist, default='Not Specified')
```

```python
"""
Rename variable
"""
summary_risk_level_sisben_2021 = df_pct 
summary_risk_level_sisben_2021
```

```python
"""
Export final dataframe with main characteristics and risk level
"""
summary_risk_level_sisben_2021.to_csv("summary_risk_level_sisben_2021.csv", sep="~", index=False, encoding = "latin1")
```

## Clustering model 2022

Please, consider the previous instructions for determine the risk level for 2022 clustering. the method is the same.

```python
GMM_2022 = pd.read_csv("df_clustered_GMM_2022.csv",delimiter="~", encoding = 'latin1' )
GMM_2022.head()
```

```python
Risk_score_2022 = GMM_2022.groupby("Cluster").mean().reset_index()
Risk_score_2022 = Risk_score_2022.drop("Cod_barrio",axis=1)
Risk_score_2022
```

```python
df_pct_22 = Risk_score_2022/Risk_score_2022[Risk_score_2022.columns].sum()
df_pct_22["Cluster"] = round(df_pct_22["Cluster"]*100/7)
df_pct_22
```

```python
df_pct_22["risk_weight"] = round(df_pct_22.mean(axis=1),2)
df_pct_22
```

```python
conditionlist_2022 = [
    (df_pct_22['risk_weight'] == 0.61 ),
    (df_pct_22['risk_weight'] == 0.22 ),
    (df_pct_22['risk_weight'] == 0.21 ),
    (df_pct_22['risk_weight'] == 0.09 ),
    (df_pct_22['risk_weight'] == 0.08), 
    (df_pct_22['risk_weight'] == 0.04)]
choicelist_2022 = ['Very-High', 'High', 'High-Mid', 'Mid', 'Mid-Low', 'Low']
```

```python
df_pct_22["risk_level"] = np.select(conditionlist_2022, choicelist_2022, default='Not Specified')
df_pct_22
```

```python
summary_risk_level_sisben_2022 = df_pct_22 
```
