```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.hist(df['chol'])
plt.show()
```

```python
plt.hist(df['chol'],bins=20)
plt.show()
```

```python
plt.hist(df['chol'],bins=5)
plt.show()
```

```python
plt.hist(df['chol'],bins=20,histtype='step')
plt.show()
```

```python
plt.hist(df['chol'],bins=20,orientation='horizontal')
plt.show()
```

```python
plt.hist(df['chol'],bins=20,alpha=0.5)
plt.hist(df['trestbps'],bins=10,alpha=0.5)
plt.show()
```

```python
plt.hist(df['chol'],bins=20,alpha=0.5,label='chol')
plt.hist(df['trestbps'],bins=10,alpha=0.5,label='trestbps')
plt.legend()
```
