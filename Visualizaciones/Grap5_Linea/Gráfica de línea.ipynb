```python
cemex = pd.read_csv('data/linea/CEMEXCPO.MX.csv',index_col=0)
bimbo = pd.read_csv('data/linea/BIMBOA.MX.csv',index_col = 0)
cemex.head()
```

```python
cemex.index = pd.to_datetime(cemex.index)
bimbo.index = pd.to_datetime(bimbo.index)
cemex.head()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(cemex['Close'])
```

```python
# linestyle or ls [ '-' | '--' | '-.' | ':' | 'steps' | ...]
# marker [ '+' | ',' | '.' | '1' | '2' | '3' | '4' ]
plt.plot(cemex['Close'],'b-',linewidth=1)
```

```python
plt.plot(cemex['Close']      ,'g-',alpha = 1.0, label='a:1.0')
plt.plot(cemex['Close'] * 2  ,'g-',alpha = 0.5, label='a:0.5')
plt.plot(cemex['Close'] *1/2 ,'g-',alpha = 0.2, label='a:0.1')
plt.legend()
```

```python
plt.plot(cemex['Low'],label='Low')
plt.plot(cemex['High'],label='High')
plt.legend()
```

```python
plt.plot(cemex['Low'],label='LowCemex')
plt.plot(cemex['High'],label='HighCemex')
plt.plot(bimbo['High'],label='HighBimbo')
plt.legend()
```

```python
plt.figure()
plt.subplot(211)
plt.plot(cemex['Low'],label='Low')
plt.plot(cemex['High'],label='High')
plt.legend()
plt.subplot(212)
plt.plot(bimbo['Low'],label='Low')
plt.plot(bimbo['High'],label='High')
plt.legend()
```

```python
plt.figure()
ax1 = plt.subplot(211)
plt.plot(cemex['Low'],label='Low')
plt.plot(cemex['High'],label='High')
plt.title('Cemex')
plt.tight_layout()
plt.legend()
ax2 = plt.subplot(212)
plt.plot(bimbo['Low'],label='Low')
plt.plot(bimbo['High'],label='High')
plt.title('Bimbo')
plt.tight_layout()
plt.legend()
```

```python
plt.figure()
ax1 = plt.subplot(211)
plt.plot(cemex['Low'],'r--',label='Low',alpha=0.3)
plt.plot(cemex['High'],'b--',label='High',alpha=0.3)
plt.plot(cemex['Close'],'g-',label='Close',alpha=1)
plt.title('Cemex')
#plt.xlim('2019-05', '2019-06')
plt.ylim(7.5, 9.5)
plt.xticks(rotation=30)
plt.tight_layout()
plt.legend()
ax2 = plt.subplot(212)
plt.plot(bimbo['Low'],'r--',label='Low',alpha=0.3)
plt.plot(bimbo['High'],'b--',label='High',alpha=0.3)
plt.plot(bimbo['Close'],'g-',label='Close',alpha=1)
plt.title('Bimbo')
#plt.xlim('2019-05', '2019-06')
plt.ylim(35, 45)
plt.xticks(rotation=30)
plt.tight_layout()
plt.legend()
plt.savefig("graficadelinea.png",dpi=300)
```
